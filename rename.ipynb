```python
import os

directory = r"D:\FCAI\plain_coupon\infared_data"

for folder_name in os.listdir(directory):
    folder_path = os.path.join(directory, folder_name)
    
    # Only process directories matching the expected pattern
    if not os.path.isdir(folder_path):
        continue
    
    parts = folder_name.split("_")
    
    # Expected format: Phi_XX_CR_YYYY (4 parts)
    if len(parts) != 4 or parts[0].lower() != "phi" or parts[2].lower() != "cr":
        print(f"Skipping (unexpected format): {folder_name}")
        continue
    
    phi_value = parts[1]   # e.g. '09', '085', '118'
    cr_value  = parts[3]   # e.g. '500', '1400', '2300'
    
    new_name = f"cr_{cr_value}_phi_{phi_value}"
    new_path = os.path.join(directory, new_name)
    
    print(f"Renaming: {folder_name}  →  {new_name}")
    os.rename(folder_path, new_path)

print("\nDone!")
```
