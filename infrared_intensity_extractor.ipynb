# Infrared Intensity Point Extractor
**Workflow:**
1. **Cell 1 – Config**: Set your base data path.
2. **Cell 2 – Window 1 Point Selection**: Click 4 points on the Window 1 image.
3. **Cell 3 – Window 2 Point Selection**: Click 4 *different* points on the Window 2 image.
4. **Cell 4 – Preview**: Verify both sets of points on the example case.
5. **Cell 5 – Extract & Export**: Runs all cases and saves `intensity_summary.xlsx`.

```python
# ── Cell 1: Configuration ──────────────────────────────────────────────────────
import os
import numpy as np
import matplotlib
matplotlib.use('TkAgg')   # interactive clicks require TkAgg backend
import matplotlib.pyplot as plt

BASE_DIR     = r"D:\FCAI\plain_coupon\infared_data"
EXAMPLE_CASE = "cr_500_phi_09"   # used for interactive point selection
HEADER_LINES = 9                  # metadata lines to skip in each CSV
N_POINTS     = 4

W1_CSV = "window_1_corrected_interp.csv"
W2_CSV = "window_2_corrected_interp.csv"

def load_csv(path):
    return np.genfromtxt(path, delimiter=',', skip_header=HEADER_LINES)

# Storage for both point groups (populated in Cells 2 & 3)
points_w1 = []   # 4 (row, col) for window_1
points_w2 = []   # 4 (row, col) for window_2

print("Config loaded.")
print(f"Base dir : {BASE_DIR}")
print(f"Example  : {EXAMPLE_CASE}")
```

```python
# ── Cell 2: Select 4 points on Window 1 ───────────────────────────────────────
# Click exactly 4 locations, then close the window.

points_w1.clear()

csv_path = os.path.join(BASE_DIR, EXAMPLE_CASE, "time_average", W1_CSV)
data_w1  = load_csv(csv_path)
print(f"Loaded: {csv_path}  |  shape: {data_w1.shape}")

fig, ax = plt.subplots(figsize=(8, 7))
im = ax.imshow(data_w1, cmap='inferno', origin='upper')
plt.colorbar(im, ax=ax, label='Intensity (Counts)')
ax.set_title(f"WINDOW 1 — click {N_POINTS} points  ·  {EXAMPLE_CASE}\n"
             "(close window when done)")

def onclick_w1(event):
    if event.inaxes != ax or len(points_w1) >= N_POINTS:
        return
    col = max(0, min(int(round(event.xdata)), data_w1.shape[1] - 1))
    row = max(0, min(int(round(event.ydata)), data_w1.shape[0] - 1))
    points_w1.append((row, col))
    n = len(points_w1)
    ax.plot(col, row, 'w+', markersize=14, markeredgewidth=2)
    ax.text(col + 2, row - 2, f"P{n}", color='white', fontsize=10, fontweight='bold')
    fig.canvas.draw()
    print(f"  W1-P{n}: row={row:>4d}, col={col:>4d}  intensity={data_w1[row, col]:.2f}")
    if n == N_POINTS:
        ax.set_title("All 4 Window-1 points selected — close this window to continue.")
        fig.canvas.draw()

fig.canvas.mpl_connect('button_press_event', onclick_w1)
plt.tight_layout()
plt.show()

print(f"\n=== Window 1 points ({len(points_w1)}/{N_POINTS}) ===")
for i, (r, c) in enumerate(points_w1, 1):
    print(f"  P{i}: row={r:>4d}, col={c:>4d}")
```

```python
# ── Cell 3: Select 4 points on Window 2 ───────────────────────────────────────
# These are INDEPENDENT from the Window 1 points — click wherever you need.

points_w2.clear()

csv_path = os.path.join(BASE_DIR, EXAMPLE_CASE, "time_average", W2_CSV)
data_w2  = load_csv(csv_path)
print(f"Loaded: {csv_path}  |  shape: {data_w2.shape}")

fig, ax = plt.subplots(figsize=(8, 7))
im = ax.imshow(data_w2, cmap='inferno', origin='upper')
plt.colorbar(im, ax=ax, label='Intensity (Counts)')
ax.set_title(f"WINDOW 2 — click {N_POINTS} points  ·  {EXAMPLE_CASE}\n"
             "(close window when done)")

def onclick_w2(event):
    if event.inaxes != ax or len(points_w2) >= N_POINTS:
        return
    col = max(0, min(int(round(event.xdata)), data_w2.shape[1] - 1))
    row = max(0, min(int(round(event.ydata)), data_w2.shape[0] - 1))
    points_w2.append((row, col))
    n = len(points_w2)
    ax.plot(col, row, 'c+', markersize=14, markeredgewidth=2)   # cyan markers for W2
    ax.text(col + 2, row - 2, f"P{n}", color='cyan', fontsize=10, fontweight='bold')
    fig.canvas.draw()
    print(f"  W2-P{n}: row={row:>4d}, col={col:>4d}  intensity={data_w2[row, col]:.2f}")
    if n == N_POINTS:
        ax.set_title("All 4 Window-2 points selected — close this window to continue.")
        fig.canvas.draw()

fig.canvas.mpl_connect('button_press_event', onclick_w2)
plt.tight_layout()
plt.show()

print(f"\n=== Window 2 points ({len(points_w2)}/{N_POINTS}) ===")
for i, (r, c) in enumerate(points_w2, 1):
    print(f"  P{i}: row={r:>4d}, col={c:>4d}")
```

```python
# ── Cell 4: Preview — verify both point sets on the example case ───────────────
assert len(points_w1) == N_POINTS, f"Need {N_POINTS} W1 points, got {len(points_w1)}. Re-run Cell 2."
assert len(points_w2) == N_POINTS, f"Need {N_POINTS} W2 points, got {len(points_w2)}. Re-run Cell 3."

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Window 1
ax1 = axes[0]
d1  = load_csv(os.path.join(BASE_DIR, EXAMPLE_CASE, "time_average", W1_CSV))
im1 = ax1.imshow(d1, cmap='inferno', origin='upper')
plt.colorbar(im1, ax=ax1, label='Intensity')
ax1.set_title(f"{EXAMPLE_CASE}\n{W1_CSV}")
for i, (r, c) in enumerate(points_w1, 1):
    ax1.plot(c, r, 'w+', markersize=14, markeredgewidth=2)
    ax1.text(c + 2, r - 2, f"P{i}", color='white', fontsize=9, fontweight='bold')

# Window 2
ax2 = axes[1]
d2  = load_csv(os.path.join(BASE_DIR, EXAMPLE_CASE, "time_average", W2_CSV))
im2 = ax2.imshow(d2, cmap='inferno', origin='upper')
plt.colorbar(im2, ax=ax2, label='Intensity')
ax2.set_title(f"{EXAMPLE_CASE}\n{W2_CSV}")
for i, (r, c) in enumerate(points_w2, 1):
    ax2.plot(c, r, 'c+', markersize=14, markeredgewidth=2)
    ax2.text(c + 2, r - 2, f"P{i}", color='cyan', fontsize=9, fontweight='bold')

plt.suptitle("Preview — white markers = W1 points  |  cyan markers = W2 points", fontsize=12)
plt.tight_layout()
plt.show()
print("If any points look wrong, re-run Cell 2 (W1) or Cell 3 (W2) to reselect.")
```

```python
# ── Cell 5: Extract intensities for ALL cases and export to Excel ──────────────
import pandas as pd
import re
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter

assert len(points_w1) == N_POINTS, "Run Cell 2 first."
assert len(points_w2) == N_POINTS, "Run Cell 3 first."

# ── Discover all case folders ──────────────────────────────────────────────────
case_pattern = re.compile(r'^cr_(\d+)_phi_(\d+)$')
cases = sorted(
    [d for d in os.listdir(BASE_DIR)
     if os.path.isdir(os.path.join(BASE_DIR, d)) and case_pattern.match(d)]
)
print(f"Found {len(cases)} case folders.")

# ── Build records ──────────────────────────────────────────────────────────────
records = []
missing = []

for case in cases:
    m        = case_pattern.match(case)
    cr       = int(m.group(1))
    phi_str  = m.group(2)
    phi      = int(phi_str) / 100.0 if len(phi_str) == 3 else int(phi_str) / 10.0
    row_data = {"Case": case, "CR (slpm)": cr, "Phi": phi}

    # Window 1 — uses points_w1 exclusively
    p1 = os.path.join(BASE_DIR, case, "time_average", W1_CSV)
    if os.path.isfile(p1):
        d = load_csv(p1)
        for pt_idx, (r, c) in enumerate(points_w1, 1):
            row_data[f"W1_P{pt_idx}"] = (
                float(d[r, c]) if (0 <= r < d.shape[0] and 0 <= c < d.shape[1]) else None
            )
    else:
        missing.append(p1)
        for pt_idx in range(1, N_POINTS + 1):
            row_data[f"W1_P{pt_idx}"] = None

    # Window 2 — uses points_w2 exclusively
    p2 = os.path.join(BASE_DIR, case, "time_average", W2_CSV)
    if os.path.isfile(p2):
        d = load_csv(p2)
        for pt_idx, (r, c) in enumerate(points_w2, 1):
            row_data[f"W2_P{pt_idx}"] = (
                float(d[r, c]) if (0 <= r < d.shape[0] and 0 <= c < d.shape[1]) else None
            )
    else:
        missing.append(p2)
        for pt_idx in range(1, N_POINTS + 1):
            row_data[f"W2_P{pt_idx}"] = None

    records.append(row_data)

df = pd.DataFrame(records)

# ── Point coordinate reference sheet ──────────────────────────────────────────
coord_rows = []
for pt_idx, (r, c) in enumerate(points_w1, 1):
    coord_rows.append({"Window": "Window 1", "Point": f"P{pt_idx}",
                       "Row (pixel)": r, "Col (pixel)": c})
for pt_idx, (r, c) in enumerate(points_w2, 1):
    coord_rows.append({"Window": "Window 2", "Point": f"P{pt_idx}",
                       "Row (pixel)": r, "Col (pixel)": c})
df_coords = pd.DataFrame(coord_rows)

# ── Save to Excel ──────────────────────────────────────────────────────────────
out_path = os.path.join(BASE_DIR, "intensity_summary.xlsx")
with pd.ExcelWriter(out_path, engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name="Intensity Data", index=False)
    df_coords.to_excel(writer, sheet_name="Point Coordinates", index=False)

# ── Formatting ─────────────────────────────────────────────────────────────────
wb = load_workbook(out_path)

HEADER_FILL = PatternFill("solid", start_color="1F4E79", end_color="1F4E79")
W1_FILL     = PatternFill("solid", start_color="D6E4F0", end_color="D6E4F0")  # light blue
W2_FILL     = PatternFill("solid", start_color="D5F5E3", end_color="D5F5E3")  # light green
HEADER_FONT = Font(name="Arial", bold=True, color="FFFFFF", size=10)
BODY_FONT   = Font(name="Arial", size=10)
CENTER      = Alignment(horizontal="center", vertical="center")
thin        = Side(style='thin', color='CCCCCC')
BORDER      = Border(left=thin, right=thin, top=thin, bottom=thin)

for sheet_name in ["Intensity Data", "Point Coordinates"]:
    ws = wb[sheet_name]
    for cell in ws[1]:
        cell.font      = HEADER_FONT
        cell.fill      = HEADER_FILL
        cell.alignment = CENTER
        cell.border    = BORDER
    col_headers = [c.value for c in ws[1]]
    for row in ws.iter_rows(min_row=2):
        for cell in row:
            cell.font      = BODY_FONT
            cell.alignment = CENTER
            cell.border    = BORDER
            if sheet_name == "Intensity Data":
                hdr = col_headers[cell.column - 1]
                if isinstance(hdr, str) and hdr.startswith("W1_"):
                    cell.fill = W1_FILL
                elif isinstance(hdr, str) and hdr.startswith("W2_"):
                    cell.fill = W2_FILL
                if isinstance(cell.value, float):
                    cell.number_format = '0.00'
    for col in ws.columns:
        max_len = max((len(str(c.value)) for c in col if c.value), default=8)
        ws.column_dimensions[get_column_letter(col[0].column)].width = min(max_len + 4, 25)
    ws.freeze_panes = "A2"

# Friendly column header names
ws = wb["Intensity Data"]
rename = {f"W{w}_P{p}": f"Win{w} P{p}" for w in [1, 2] for p in range(1, N_POINTS + 1)}
for cell in ws[1]:
    if cell.value in rename:
        cell.value = rename[cell.value]

wb.save(out_path)

# ── Summary ───────────────────────────────────────────────────────────────────
print(f"\n✅  Done!  {len(records)} cases processed.")
if missing:
    print(f"⚠️  {len(missing)} missing CSV files (recorded as blank):")
    for p in missing:
        print(f"   {p}")
print(f"\n📄  Output saved to:\n    {out_path}")
df.head(10)
```
