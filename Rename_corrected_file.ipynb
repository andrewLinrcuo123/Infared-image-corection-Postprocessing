```python
"""
Paste this entire cell into your Jupyter notebook and run it.
Change NEW_NAME to whatever you want the file to be called.
"""

import os

# ── Configuration ─────────────────────────────────────────────────────────────
BASE_DIR      = r"D:\FCAI\plain_coupon\infared_data"
CURRENT_NAME  = "corrected.csv"
NEW_NAME      = "window_1_corrected.csv"       # <-- change this to your desired name
OUTPUT_SUBDIR = "time_average"
SKIP_FOLDERS  = {"time_average", "time_averaged_plot_intensity"}
# ──────────────────────────────────────────────────────────────────────────────

# Auto-append .csv if missing
if not NEW_NAME.lower().endswith(".csv"):
    NEW_NAME = NEW_NAME + ".csv"
    print(f"Note: '.csv' appended → '{NEW_NAME}'")

if NEW_NAME == CURRENT_NAME:
    print(f"ERROR: new name is the same as current name ('{CURRENT_NAME}'). Nothing done.")
else:
    base = os.path.normpath(BASE_DIR)

    entries   = sorted(os.listdir(base))
    case_dirs = [
        os.path.join(base, e) for e in entries
        if os.path.isdir(os.path.join(base, e)) and e not in SKIP_FOLDERS
    ]

    print("=" * 62)
    print(f"  Renaming  '{CURRENT_NAME}'  →  '{NEW_NAME}'")
    print("=" * 62)

    found_list   = []
    missing_list = []

    for case_dir in case_dirs:
        case_name = os.path.basename(case_dir)
        src_path  = os.path.join(case_dir, OUTPUT_SUBDIR, CURRENT_NAME)
        dst_path  = os.path.join(case_dir, OUTPUT_SUBDIR, NEW_NAME)

        if not os.path.isfile(src_path):
            print(f"  [MISSING] {case_name}  — '{CURRENT_NAME}' not found")
            missing_list.append(case_name)
            continue

        if os.path.isfile(dst_path):
            print(f"  [SKIP]    {case_name}  — '{NEW_NAME}' already exists")
            missing_list.append(case_name)
            continue

        os.rename(src_path, dst_path)
        print(f"  [OK]      {case_name}  →  {NEW_NAME}")
        found_list.append(case_name)

    print("\n" + "=" * 62)
    print("  SUMMARY")
    print("=" * 62)
    print(f"  Renamed : {len(found_list)}")
    print(f"  Skipped : {len(missing_list)}")
    if missing_list:
        print("\n  Cases where rename was skipped:")
        for c in missing_list:
            print(f"    x  {c}")
    print("=" * 62)
```
